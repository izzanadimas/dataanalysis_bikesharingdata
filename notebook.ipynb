### Cleaning Data

#### Cleaning Data: Bike Sharing Data by Day

Berdasarkan hasil assessing data, didapatkan hanya ada kesalahan pada kolom **dteday** yang masih bertipe data **object**. Untuk itu, tipe data kolom **dteday** perlu diganti menjadi **datetime**.

```python
# Mengubah tipe data
byday_df_copy = byday_df.copy() #Membuat duplikat data supaya data lama tetap tersimpan

# Mengubah tipe data
byday_df_copy["dteday"] = pd.to_datetime(byday_df_copy["dteday"])

# Memeriksa kembali info data
byday_df_copy.info()
```

Dapat dilihat bahwa tipe data dari kolom "dteday" sudah berubah menjadi **datetime**. Dengan begitu, tahap _cleanning_ data sudah selesai.

Untuk mempermudah pengategorian dan analisis, keterangan cuaca pada kolom **weathersit** akan diubah menjadi data kategori dan keterangan tahun pada kolom **yr**, di mana sebelumnya data hanya dikatgorikan menggunakan angka.

```python
# Mengubah keterangan pada tabel Bike Sharing Data in Hour kolom weathersit
for index, row in byday_df_copy.iterrows():
    if row['season'] == 1:
        byday_df_copy.loc[index, 'season'] = 'Spring'
    elif row['season'] == 2:
        byday_df_copy.loc[index, 'season'] = 'Summer'
    elif row['season'] == 3:
        byday_df_copy.loc[index, 'season'] = 'Fall'
    elif row['season'] == 4:
        byday_df_copy.loc[index, 'season'] = 'Winter'

# Mengubah keterangan pada tabel Bike Sharing Data in Hour kolom weathersit
for index, row in byday_df_copy.iterrows():
    if row['weathersit'] == 1:
        byday_df_copy.loc[index, 'weathersit'] = 'Clear'
    elif row['weathersit'] == 2:
        byday_df_copy.loc[index, 'weathersit'] = 'Cloudy/Mist'
    elif row['weathersit'] == 3:
        byday_df_copy.loc[index, 'weathersit'] = 'Light rain/Snow'
    elif row['weathersit'] == 4:
        byday_df_copy.loc[index, 'weathersit'] = 'Heavy rain/Snow'

# Mengubah keterangan pada tabel Bike Sharing Data in Hour kolom yr
for index, row in byday_df_copy.iterrows():
    if row['yr'] == 0:
        byday_df_copy.loc[index, 'yr'] = 2011
    elif row['yr'] == 1:
        byday_df_copy.loc[index, 'yr'] = 2012

byday_df_copy
```

Terlihat bahwa kolom **season**, **yr**, dan **weathersit** sudah dimodifikasi sehingga dapat lebih mudah untuk dibaca.

## Exploratory Data Analysis (EDA)

### Explore the bike sharing business performance

Untuk mengetahui bagaimana progres bisnis peminjaman sepeda tersebut, dapat dibuat **pivot table** untuk mengetahui berapa jumlah sepeda yang dipinjam setiap harinya.

```python
byhour_df_copy.groupby(by="dteday").agg({
    'casual': 'sum',
    'registered': 'sum',
    'cnt': 'sum',
})
```

Mengacu pada **pivot table** tersebut, ternyata terdapat terlalu banyak baris data jika data direkapitulasi untuk untuk banyak peminjaman per harinya. Untuk itu, perlu dibuat tabel yang lebih ringkas, yaitu tabel rekapitulasi banyak peminjaman sepeda setiap bulannya.

```python
monthly_sharing_df = byhour_df_copy.resample(rule='ME', on='dteday').agg({
    "casual": "sum",
    "registered": "sum"
})
monthly_sharing_df.index = monthly_sharing_df.index.strftime('%Y-%m')
monthly_sharing_df = monthly_sharing_df.reset_index()
monthly_sharing_df.rename(columns={
    "casual": "casual",
    "registered": "registered"
}, inplace=True)
monthly_sharing_df
```

Dengan menggunakan data rekapitulasi banyak peminjaman sepeda tiap bulan, pergerakan/progres bisnis jauh lebih mudah untuk dinilai.

### Explore how the season affects the number of bikes borrowed

Untuk mengetahui rata-rata banyak peminjaman sepeda tiap musim, digunakan **pivot table** untuk melihat rekapitulasi rata-rata peminjaman sepeda pada tiap musim di tahun 2011 dan 2012.

```python
byday_df_copy.groupby(by=["yr","season"]).agg({
    'cnt': 'mean',
})
```

Setiap musim pastinya memiliki suhu rata-rata yang berbeda. Supaya faktor perbedaan jumlah peminjaman setiap musim dapat diketahui dengan lebih jelas, dapat dilihat suhu rata-rata tiap musim.

```python
byday_df_copy.groupby(by=["yr","season"]).agg({
    'temp': 'mean',
})
```

Berdasarkan kedua hal tersebut, didapatkan asumsi bahwa suhu tiap musim mempengaruhi banyak peminjam sepeda. Oleh karena itu, perlu dilakukan validasi hubungan dengan mengidentifikasi korelasi tiap data pada tabel menggunakan fungsi **corr()**.

```python
byday_df_copy.corr(numeric_only=True)
```

Berdasarkan hasil identifikasi korelasi, didapatkan bahwa temperatur/suhu memiliki korelasi positif dengan banyak peminjam sepeda. Dengan begitu, asumsi bahwa musim dapat mempengaruhi banyak peminjam sepeda dapat diterima.

## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana performa usaha peminjaman sepeda pada tahun 2011 dan 2012?

```python
# Mengatur ukuran
plt.figure(figsize=(28, 7))

# Plot
plt.plot(monthly_sharing_df['dteday'], monthly_sharing_df['registered'], marker="o", color='blue', label="Registered")
plt.plot(monthly_sharing_df['dteday'], monthly_sharing_df['casual'], marker="o", color='green', label="Casual")
plt.xlabel('Year-Month')
plt.ylabel('Number of borrowed bikes')
plt.title('Performa Bisnis Peminjaman Sepeda Tahun 2011 - 2012', size=15)
plt.grid(True)
plt.legend(title="Customer category", loc='upper left')

# Menampilkan plot
plt.show()
```

### Pertanyaan 2:
Bagaimana musim mempengaruhi peminjaman sepeda?

```python
# Data
season_year_df = byday_df_copy.groupby(["yr","season"])['cnt'].mean().reset_index()
season_temp_df = byday_df_copy.groupby(["yr","season"])['temp'].mean().reset_index()

# Mengatur ukuran
plt.figure(figsize=(20, 6))

# Plot pertama: Rata-rata peminjaman sepeda tiap musim pada tahun 2011 dan 2012
plt.subplot(1, 2, 1)
sns.barplot(data=season_year_df, x="yr", y="cnt", hue="season", errorbar=None)
plt.xlabel("Tahun")
plt.ylabel("Jumlah Peminjaman Sepeda")
plt.title("Rata-Rata Peminjaman Sepeda Tiap Musim")

# Plot kedua: Rata-rata suhu tiap musim pada tahun 2011 dan 2012
plt.subplot(1, 2, 2)
sns.barplot(data=season_temp_df, x="yr", y="temp", hue="season", errorbar=None)
plt.xlabel("Tahun")
plt.ylabel("Rata-Rata Suhu (Celsius)")
plt.title("Rata-Rata Suhu Tiap Musim")

# Menampilkan plot
plt.tight_layout()
plt.show()

```
